Write a Python code to create a word cloud from the following words.

LLM, MCP, AI Agent, A2A, Agentic AI, Agent, Multi Agent System, Actions, Tools, Agentspace, Agent Builder, Agent Engine, Agent Garden, LangChain, LangGraph. Show only the exact words presented above

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# List of words provided
word_list = [
    "LLM", "MCP", "AI_Agent", "A2A", "Agentic_AI", "Agent",
    "Multi_Agent_System", "Actions", "Tools", "Agentspace",
    "Agent_Builder", "Agent_Engine", "LangChain", "LangGraph"
]

# Join the words into a single string, separated by spaces
# This is how the WordCloud library expects its input
text = " ".join(word_list)

# --- Create the WordCloud object ---
# You can customize various parameters here (e.g., width, height, background_color)
# collocations=False ensures that the library doesn't try to group words into common phrases
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    collocations=False, # Process words individually
    collocation_threshold=200, # Effectively disables collocations for distinct phrases
    stopwords=None # No stopwords to remove as we want all given words
).generate(text)

# --- Display the generated image using matplotlib ---
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Hide the axes
plt.tight_layout(pad=0)
plt.show()
```
